## `ORDER BY` — putting rows in a deliberate order

`ORDER BY` sorts your results by one (or more) column. It comes **after** `WHERE` and **before** `LIMIT`:

```
SELECT [columns]
FROM [table]
WHERE [condition]
ORDER BY [column]
LIMIT [count]
```

Try sorting princesses by year, oldest first:

````python
%%bigquery
SELECT *
FROM disney_lessons.princesses
ORDER BY year
````

All 16 rows, but now in order — Snow White (1937) at the top, Asha (2023) at the bottom.

`ORDER BY` defaults to **ascending** order: smallest-to-largest for numbers, A-to-Z for text, oldest-to-newest for dates.

## `DESC` — descending order

To flip the order, add `DESC` after the column name. *Newest* princesses first:

````python
%%bigquery
SELECT *
FROM disney_lessons.princesses
ORDER BY year DESC
````

Same 16 rows, but now Asha (2023) is at the top and Snow White (1937) at the bottom.

(For "ascending" you can also write `ASC` explicitly — `ORDER BY year ASC` — but it's the default, so people usually leave it off.)

## Sorting by text

You can sort by any column, not just numbers. Try alphabetically by name:

````python
%%bigquery
SELECT name, movie, year
FROM disney_lessons.princesses
ORDER BY name
````

Anna, Ariel, Asha, Aurora… all the way down to Tiana. Text sorts alphabetically by default.

## Sorting by multiple columns

You can sort by more than one column. The first column is the primary sort; the second one breaks ties.

For example: sort by movie alphabetically, and within each movie sort by year:

````python
%%bigquery
SELECT name, movie, year
FROM disney_lessons.princesses
ORDER BY movie, year
````

Notice **Frozen** has two princesses (Anna and Elsa) — they appear together because they share a movie. Each column can have its own direction:

```sql
ORDER BY movie ASC, year DESC
```

would sort movies A→Z, but within a movie show newer entries first.

## The "top N" pattern

`ORDER BY` + `LIMIT` is one of the most useful patterns in SQL. *"Show me the N smallest/biggest/newest/oldest..."* — it's everywhere.

The 3 newest princesses:

````python
%%bigquery
SELECT name, movie, year
FROM disney_lessons.princesses
ORDER BY year DESC
LIMIT 3
````

3 rows: Asha (2023), Raya (2021), and Moana (2016).

This pattern works for *anything*: the 5 most expensive products, the 10 most recent customers, the 3 longest movies. We'll use it constantly.

## `DISTINCT` — only unique values

Sometimes you want to know what *unique* values exist in a column, without the duplicates.

In Lesson 01, when you ran `SELECT movie FROM ... LIMIT 16`, you saw "Frozen" appear twice (once for Anna, once for Elsa). `DISTINCT` removes the duplicates.

````python
%%bigquery
SELECT DISTINCT movie
FROM disney_lessons.princesses
ORDER BY movie
````

**15 rows** — instead of 16. Frozen appears just once now.

`DISTINCT` goes right after `SELECT`, and applies to **whichever columns you select**. Combine it with `ORDER BY` for a clean alphabetical list.

## `DISTINCT` with multiple columns

When you `SELECT DISTINCT` on multiple columns, "unique" means *unique combinations* of those columns:

````python
%%bigquery
SELECT DISTINCT movie, year
FROM disney_lessons.princesses
ORDER BY year
````

Still 15 rows — same as `SELECT DISTINCT movie` — because `(movie, year)` is just as unique as `movie` alone.

If two rows had the same movie but different years (which can happen for re-releases), they'd appear as separate rows here.
