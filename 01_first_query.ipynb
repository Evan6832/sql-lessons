## A reminder about SQL vs. Colab plumbing

You're learning **SQL** — the actual language for asking databases questions. SQL is universal: you'll use it in BigQuery, the BigQuery web Console, BigQuery Studio, internal Google data tools at our company, and lots of other places.

You're using **Colab** as the place where you type and run that SQL. Colab is just one tool among many. It has a small "plumbing" line at the top of every SQL cell that looks like this:

```
%%bigquery
```

**That line is not SQL.** It's Colab's way of saying *"treat the rest of this cell as a SQL query and send it to BigQuery."*

When you eventually write SQL in **other tools** — like the BigQuery web Console, BigQuery Studio, or any internal data platform at work — you will **drop the `%%bigquery` line** and write the SQL exactly the same. Everything *below* that line is real, portable, universal SQL.

You'll see this reminder a few times in this lesson. The point is: **don't memorize `%%bigquery`. It's tool-specific noise. Focus on the SQL.**

## What the heck just happened — line by line

The cell you ran has four lines, and **only three of them are real SQL.** Let's go through each.

### Line 1: `%%bigquery`

**This is not SQL.** This is Colab's plumbing line — it tells Colab "take everything below this and run it as a SQL query against BigQuery." When you write SQL in other tools (BigQuery Console, BigQuery Studio, internal data tools at work), you'll drop this line entirely.

(Yes — same `%%bigquery` reminder again. It's worth repeating.)

### Line 2: `SELECT *`

This is the first piece of real SQL.

- **`SELECT`** is the command for "give me data."
- **`*`** (the asterisk) is shorthand for "every column."

So `SELECT *` means: *"Give me every column."*

(In a moment we'll switch from `*` to picking specific columns by name — that's almost always what you actually want.)

### Line 3: `FROM disney_lessons.princesses`

- **`FROM`** says where to look.
- **`disney_lessons.princesses`** is the address of the table. Read the dot as "the princesses table inside the disney_lessons dataset."

So `FROM disney_lessons.princesses` means: *"...from the princesses table in the disney_lessons dataset."*

### Line 4: `LIMIT 5`

**`LIMIT`** caps how many rows you get back. `LIMIT 5` means: *"...but stop after 5 rows."*

### Putting it together

The whole query reads as: *"Give me every column from the princesses table — but only the first 5 rows."*

That's the entire query. The thing that came back below it — the small table of 5 princesses — is the **result**. The result of any SQL query is itself a table, always.

## Picking specific columns

`SELECT *` is fine for a quick peek, but most of the time you only want **specific columns**. Instead of `*`, you list the column names you want, separated by commas.

Run the cell below.

````python
%%bigquery
SELECT name, movie, year
FROM disney_lessons.princesses
LIMIT 10
````

You should now see 10 rows with only **3 columns** — `name`, `movie`, `year`. The `home` column is gone, because you didn't ask for it.

This is one of the most important muscles to build in SQL: **only ask for what you need.** Real tables at work might have hundreds of columns. Asking for `*` on a real table is like asking for the entire encyclopedia when you wanted one paragraph.

> Reminder: `%%bigquery` on the top line is still just Colab plumbing. The real SQL is the three lines below it. In other tools, you'd just write:
> ```sql
> SELECT name, movie, year
> FROM disney_lessons.princesses
> LIMIT 10
> ```
> and that's it.

## About `LIMIT`

The princesses table has only 16 rows total, so `LIMIT` doesn't matter much here. But it's a habit worth developing now.

Real-world tables you'll query at work might have **millions of rows**. Without a `LIMIT`, your query would try to return all of them, which:

- Takes much longer
- Can be impossible to read
- Uses up more of your free monthly query quota

While you're learning, **always include `LIMIT`** when you're exploring a new table. Start with `LIMIT 10` to see the shape of the data, then remove it later only if you really need every row.

> **One subtlety:** `LIMIT` controls *how many* rows come back, but it doesn't control *which* ones. There's no guarantee that "the first 5" you get are the same 5 you'd get if you ran the same query again. The order isn't defined unless you explicitly tell SQL how to sort. You'll learn the `ORDER BY` clause for that in Lesson 03.

## Read-only, by design

You have **read-only** access to the Disney dataset. You can ask any question you can think of, and you cannot accidentally break or change anything.

If you ever try to write a query that *would* modify data (like `INSERT`, `UPDATE`, or `DELETE`), BigQuery will politely refuse with an error message — nothing happens. Treat it as a friendly nudge. We won't be doing those in this course anyway.

This means you can experiment freely. **Try things.** Change `LIMIT 5` to `LIMIT 50`. Try column names we haven't used yet. Misspell things on purpose to see what error you get. The worst that happens is BigQuery returns a clear error message and you fix it. Errors are part of learning — they're not failures, they're feedback.
