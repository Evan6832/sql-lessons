# Lesson 00 — What is SQL?

Welcome! Before we write any code, let's get you comfortable with what SQL is, what it does, and how to think about it. This whole lesson is just reading — no exercises, no buttons to click, nothing to run. About 10–15 minutes.

By the end you should feel like SQL is a "thing I understand the shape of" — not a foreign language yet, but no longer a mystery either.

## Why are we learning this?

You probably already work with data every day. Customer dashboards. Reports in Slack. Spreadsheets you pull from various tools. All of that data lives in **databases** — big organized stores of information — and SQL is the language people use to ask those databases questions.

Right now, when you need a specific piece of data, you might:
- Ask an analyst to pull it
- Wait for a dashboard to be built
- Hunt through a spreadsheet that's almost-but-not-quite what you need

Learning SQL means **you can ask the questions yourself.** Want to know how many customers signed up in March? You can write a query and find out in 10 seconds, instead of asking someone else and waiting hours.

If you've used Excel formulas before — `=SUM()`, `=COUNTIF()`, filters, pivot tables — you already understand most of the *concepts* SQL is built on. SQL is the same set of ideas (filter, sort, count, group), expressed in text, designed to work on tables that are far too big to fit in a spreadsheet.

## The four words to know

These are the building blocks. Spend a moment with each one — they're the vocabulary the rest of the course depends on.

### 1. Database
A **database** is a collection of data, organized into one or more tables. You can think of it like a workbook in Excel — except a database can hold tables with millions or billions of rows, and many people can read from it at the same time without stepping on each other.

### 2. Table
A **table** is a grid of data, made up of **rows** and **columns**. Tables look exactly like spreadsheets:

| name  | movie                  | year | home       |
|-------|------------------------|------|------------|
| Belle | Beauty and the Beast   | 1991 | Villeneuve |
| Mulan | Mulan                  | 1998 | China      |
| Moana | Moana                  | 2016 | Motunui    |

That's a tiny snippet of the first table you'll work with — a list of Disney princesses.

- Each **row** is one record. In this table, each row represents one princess.
- Each **column** holds one kind of information. The `year` column always holds a year. The `home` column always holds a place name. Columns are consistent: a column never has years in some rows and names in others.

### 3. Schema
The **schema** of a table is just the list of its columns and what type of data each one holds. The schema for our princesses table is:

| Column   | Type   | Meaning |
|----------|--------|---------|
| `name`   | text   | The princess's name |
| `movie`  | text   | The film she appears in |
| `year`   | number | Year the film was released |
| `home`   | text   | Where she's from |

That **type** column is important. **Text** columns hold words and characters (like `"Belle"`). **Number** columns hold actual numbers you can do math on (like `1991`). There are also types for **dates**, **true/false** values, and others.

The database enforces these types — you can't accidentally put a name where a year should go. Think of the schema as the contract: *every row in this table will have these columns, of these types.*

### 4. Query
A **query** is a question you ask the database, written in SQL. The database reads your query, looks at the data, and returns the answer as another (usually smaller) table.

Important idea: **the result of a SQL query is itself a table.** SQL takes tables in, gives tables out. Always.

## The shape of every query

Every SQL query you'll write follows the same basic shape:

> **"Give me _these columns_ FROM _this table_ WHERE _these conditions_."**

That's it. Sorting, grouping, joining — those are decorations on top, but every query starts here.

In real SQL, that shape looks like this:

```sql
SELECT name, movie
FROM disney_lessons.princesses
WHERE year >= 2000
```

Read that out loud as English: *"Give me the `name` and `movie` columns from the `princesses` table inside the `disney_lessons` collection, but only for rows where the `year` is 2000 or later."*

You don't need to write or memorize this yet. We'll build it up piece by piece starting in Lesson 01.

## SQL is universal — tools change, SQL stays the same

This is important: **SQL is a language, not a tool.**

Lots of different tools speak SQL — BigQuery, the BigQuery web Console, BigQuery Studio, Google Sheets' Connected Sheets feature, internal data platforms at most companies, plus countless others outside Google (Postgres, Snowflake, MySQL, etc.). Each has its own interface for typing in your query and seeing the results. **But the SQL itself is largely the same in all of them.**

We're going to use **Colab** (a free, browser-based notebook environment) as our learning interface. Colab adds a tiny bit of "tool plumbing" around your SQL — specifically, one line at the top of every query cell that says `%%bigquery`. **That line is not SQL.** It's just Colab's way of telling itself "treat the rest of this cell as a SQL query."

When you eventually write SQL in **other tools** at work — BigQuery's web console, BigQuery Studio, an internal Google data tool, anywhere — you'll just **drop the `%%bigquery` line** and write the SQL exactly the same. Everything you'll learn below that line is real, portable, universal SQL.

In other words: don't waste energy on the Colab-specific bits. The actual *SQL* is the part that transfers everywhere.

## A few quirks of SQL to know

These come up over and over, so worth getting them in your head now.

- **SQL keywords aren't case-sensitive.** `SELECT` and `select` mean exactly the same thing. By convention, people write keywords (`SELECT`, `FROM`, `WHERE`, etc.) in UPPERCASE so they stand out from your column names — but the database doesn't care.
- **SQL doesn't care about whitespace.** You can put each clause on its own line, indent however you want, or jam everything onto one line. Readability is the only reason to format. We'll always format for readability:

  ```sql
  SELECT name, movie
  FROM disney_lessons.princesses
  WHERE year >= 2000
  ```

  is exactly the same query as

  ```sql
  SELECT name,movie FROM disney_lessons.princesses WHERE year>=2000
  ```

  but the first one is much easier to read.

- **SQL has dialects.** The basics are universal, but each tool has small differences (mainly in functions for working with dates, strings, etc.). We're learning **GoogleSQL** — the dialect used by BigQuery. Most of what you learn will transfer directly to other dialects.

## A note on the data

The Disney dataset you'll be querying — `disney_lessons` — has already been set up for you in BigQuery by your project lead. You don't need to create it, load it, or manage it. You just **read** from it: you write `SELECT` queries, and BigQuery sends back rows.

Throughout this entire course, you have **read-only** access. That means you can write any query you can think of, and you cannot accidentally break, modify, or delete anything. **Experiment freely.** There is genuinely no way for you to break the data, the project, or your computer.

If you ever try a query that *would* modify data — commands like `INSERT`, `UPDATE`, or `DELETE` — the database will return a polite error and nothing will happen. We won't be doing those in this course anyway, but it's nice to know the safety net is there.
